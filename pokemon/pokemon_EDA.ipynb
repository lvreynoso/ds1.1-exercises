```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sea
```

```python
df = pd.read_csv('pokemon_utf.csv', index_col=0)
df.head()
```

```python
df.describe()
```

```python
sea.lmplot(x="Attack",y="Defense", data=df)
```

```python
sea.lmplot(x="Attack", y="Defense", data=df, fit_reg=False, hue="Stage")
```

```python
plt.xlim(0, 200)
plt.ylim(0, 200)

stats_df = df.drop(["Total", "Stage", "Legendary"], axis=1)
sea.boxplot(data=stats_df)
```

```python
plt.subplots(figsize=(20,10))

sea.set_style("whitegrid")
sea.violinplot(x="Type 1", y="Attack", data=df)
```

```python
plt.subplots(figsize=(20, 10))

pokemon_palette = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
sea.violinplot(x="Type 1", y="Attack", data=df, palette=pokemon_palette)
```

```python
plt.subplots(figsize=(20, 10))

sea.swarmplot(x="Type 1", y="Attack", data=df, palette=pokemon_palette)
```

```python
plt.figure(figsize=(20, 10))

sea.violinplot(x="Type 1", y="Attack", data=df, inner=None, palette=pokemon_palette)
sea.swarmplot(x="Type 1", y="Attack", data=df, color="k", alpha=0.7)
plt.title("Attack by Type")
```

```python
corr = stats_df.corr()
sea.heatmap(corr)
```

```python
sea.distplot(df.Attack)
```

```python
sea.countplot(x="Type 1", data=df, palette=pokemon_palette)
plt.xticks(rotation=-45)
```

```python
viz = sea.catplot(x="Type 1", y="Attack", data=df, hue="Stage", col="Stage", kind="swarm")
viz.set_xticklabels(rotation=-45)
```

```python
plt.subplots(figsize=(15,10))

sea.kdeplot(df.Attack, df.Defense)
```

```python
sea.jointplot(x="Attack", y="Defense", data=df)
```
